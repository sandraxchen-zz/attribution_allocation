```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("attribution_allocation_student_data.csv")

df.head()
```

```python
df.convert_TF.value_counts()
```

## Part 1: Attribution

```python
df_attribution = df[df['convert_TF'] == True].drop(['convert_TF', 'tier'], axis = 1)

df_attribution['id'] = np.arange(len(df_attribution))

df_attribution = pd.melt(df_attribution, id_vars=['id'], value_vars=['touch_1', 'touch_2', 'touch_3', 'touch_4', 'touch_5'],
       var_name = 'touch', value_name ='channel').sort_values(['id','touch'])

#drop nas
df_attribution =df_attribution.dropna(axis=0)

df_attribution['touch'] = df_attribution['touch'].str[-1].astype(int)

df_attribution['max_touch'] = df_attribution.groupby('id').touch.transform('max')

df_attribution['max_ind'] = np.where(df_attribution.touch == df_attribution.max_touch, 1, 0)

df_attribution.head(20)
```

```python
# last interaction
df_attribution[df_attribution['max_ind'] == 1].channel.value_counts()
```

```python
# last non-direct
df_attribution_nodirect = df_attribution[df_attribution['channel'] != 'direct']

df_attribution_nodirect['max_ind'] = np.where(df_attribution_nodirect.touch == df_attribution_nodirect.groupby('id').touch.transform('max'), 1, 0)

df_attribution_nodirect[df_attribution_nodirect['max_ind'] == 1].channel.value_counts()
```

```python
# first interaction
df_attribution[df_attribution['touch'] == '1'].channel.value_counts()
```

```python
# position based

def attribution(row):
    if row['max_touch'] == 1:
        val = 1
    elif row['max_touch'] == 2:
        val = 0.5
    elif row['touch'] == 1 or row['max_ind'] == 1:
        val = 0.4
    else:
        val = 0.2 / (int(row['max_touch']) - 2)
        
    return val
    
df_attribution['attr'] = df_attribution.apply(attribution, axis=1)

#check if I did it correctly
df_attribution['check'] = df_attribution.groupby('id').attr.transform(sum)
print("Check error: %d" % (sum(df_attribution['check'] != 1)))

df_attribution.groupby('channel').attr.agg('sum').reset_index()
```
